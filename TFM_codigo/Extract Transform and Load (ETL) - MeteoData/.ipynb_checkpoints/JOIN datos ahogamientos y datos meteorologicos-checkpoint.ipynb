```python
# Librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Datos ahogamiento
datos_ahogamiento = pd.read_csv("datos-ahogamiento.csv")

# Limpiamos las variables vacías
datos_ahogamiento = datos_ahogamiento.iloc[:, :27] # quitamos las columnas vacías de información
```

```python
datos_ahogamiento.shape
```

```python
datos_ahogamiento.head()
```

```python
# Datos meteorologicos
datos_meteorologicos = pd.read_csv("datos_meteorologicos.csv")
```

```python
datos_meteorologicos.shape
```

```python
datos_meteorologicos
```

## Creamos el JOIN

```python
# Dataframe Final 
data = pd.merge(datos_ahogamiento, datos_meteorologicos, on=['Fecha', 'Provincia'], how='left')
data.shape
```

```python
# Eliminamos la última coluna que no aporta información
data = data.iloc[:, :-1]
```

```python
# Guardamos el dataset final
data.to_csv('datos_ahogamiento_meteorologicos.csv', index=False)
```
