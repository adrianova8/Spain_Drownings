# Extraemos los datos meteorológicos por año de todas las estaciones de España

+ Una vez tenemos los datos de todas las estaciones para los años de 2013-2020, procedemos a eliminar los duplicados, ya que cuando especificamos en el código los periodos 01-01-20XX al 31-01-20XX estamos diciendo que nos coja para todos los meses 31 días y como bien sabemos hay meses que no tienen 31 días, de modo que el sistema lo que hace es coger 31 días independientemente, asi que si encontramos en el mes de febrero y especifícamos 31 días cogerá los 3 días siguientes de marzo. 


+ Una vez explicado este pequeño problema, simplemente eliminando los duplicados del dataframe nos aseguramos que no tenemos repetida más de una vez una misma fila. 

```python
# Dataframe sin duplicados
datos_meteorologicos = df.drop_duplicates()
datos_meteorologicos.head()
```

```python
# Seleccionamos aquellas variables de interés y descartamos las que no usaremos
datos_meteorologicos = datos_meteorologicos.drop(["horatmin", "horatmax", "horaracha", \
                                                 "horaPresMax", "horaPresMin"], axis = 1)
```

```python
# TODOS los datos son de tipo "object"
datos_meteorologicos.dtypes
```

```python
# Convertimos aquellos datos que son de tipo integer y fecha
numeric_columns = ["tmed", "prec", "tmin", "tmax", "velmedia", "racha", "sol", "presMax", "presMin"]
date_column = ["fecha"]

# Convertimos los datos numericos que contienen "," por ".". Además, convertimos los datos numéricos a integer
datos_meteorologicos[numeric_columns] = datos_meteorologicos[numeric_columns].applymap(lambda x: str(x).replace(',', '.')).apply(pd.to_numeric, errors = "coerce")
datos_meteorologicos["altitud"] = pd.to_numeric(datos_meteorologicos["altitud"], errors="coerce")
datos_meteorologicos["dir"] = pd.to_numeric(datos_meteorologicos["dir"], errors="coerce")

# Convertimos la variable "fecha" a "date"
datos_meteorologicos[date_column] = datos_meteorologicos[date_column].apply(pd.to_datetime, format="%Y-%m-%d", errors = "ignore")
```

```python
# Cambiamos el nombre de las variables para hacerlas más amigables
datos_meteorologicos.columns = ["Fecha", "Indicador", "Estacion", "Provincia", "Altitud", "TempMed", "Precip", \
                               "TempMin", "TempMax", "DirViento", "VelMedViento", "RachaViento", "TiempoSol", \
                               "PresionMax", "PresionMin"]
```

# SELECCION DE UNA UNICA ESTACION POR PROVINCIA

+ He seleccionado la estación que dispone de más datos, ya que además es la estación que suele representar más a la provincia

```python
# Creamos una copia del dataframe
datos_meteorologicos_copia = datos_meteorologicos.copy()
datos_meteorologicos_copia.head(2)
```

```python
# Agrupa los datos por provincia y estación meteorológica
counts = datos_meteorologicos_copia.groupby(['Provincia', 'Estacion']).size()

# Ordena los grupos por el número de registros en orden descendente
sorted_counts = counts.groupby('Provincia', group_keys=False).nlargest(1).reset_index(level=[0,1])

# Unir el dataframe original con el índice seleccionado utilizando merge
estacionProvincia = pd.merge(datos_meteorologicos_copia, sorted_counts, on=['Provincia', 'Estacion'])
```

```python
# Guardamos el dataset como CSV
estacionProvincia.to_csv('datos_meteorologicos.csv', index=False)
```
